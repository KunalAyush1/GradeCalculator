```python
import numpy as np
import pandas as pd
```

```python
print(f"Hello User!!")
print(f"Which type of service do you want to use:")
print(f"1. You have mean of class and standard deviation of class and just need to calculate your grade")
print(f"2. You have marks of all students of your class in the form of csv: ")
print(f"3. You want to manually enter marks of each and every student:")

inp =  int(input("Please choose the option to proceed: "))

```

```python
if (inp == 1):
    mean_user = float(input("Enter mean marks obtained:"))
    std_user = float(input("Enter standard deviation:"))
elif (inp ==2):
    link_csv = string(input("Paste the link of csv file:"))
elif(inp == 3):
    no_of_students = int(input("Enter the number of students in your batch:"))
    marks_array = np.zeros(no_of_students)
    for i in range(no_of_students):
        marks_array[i] = float(input(f"Enter the marks of {i+1}th student: "))
else:
    print("Invalid Input detected!!!")
        
```

```python
#Grading logic 
def grading_cutoff(avg,stdev):
    A_plus_cutoff = np.minimum((avg + 1.5*(stdev)), 91)
    A_cutoff = np.minimum((avg + 1.0*(stdev)), 82)
    B_plus_cutoff = np.minimum((avg + 0.5*(stdev)), 73)
    B_cutoff = np.minimum((avg), 64)
    C_plus_cutoff = np.minimum((avg - 0.5*(stdev)), 55)
    C_cutoff = np.minimum((avg - 1.0*(stdev)), 46)
```

```python
def grade_printer(marks):
    if(marks >= A_plus_cutoff):
        print(f"Grade a/c to your given data is A+(10)")
    elif(marks < A_plus_cutoff and marks >= A_cutoff):
        print(f"Grade a/c to your given data is A(9)")
    elif(marks < A_cutoff and marks >= B_plus_cutoff):
        print(f"Grade a/c to your given data is B+(8)")
    elif(marks < B_plus_cutoff and marks >= B_cutoff):
        print(f"Grade a/c to your given data is B(7)")
    elif(marks < B_cutoff and marks >= C_plus_cutoff):
        print(f"Grade a/c to your given data is C+(6)")
    elif(marks < C_plus_cutoff and marks >= C_cutoff):
        print(f"Grade a/c to your given data is C(5)")
    elif(marks < C_cutoff and marks >= 35):
        print(f"Grade a/c to your given data is D(4)")
    else:
        print(f"Grade a/c to your given data is unfortunately F(Fail)")
```

```python
if(inp == 1):
    marks_of_user = float(input("Enter your marks to calculate your grade:"))
    grading_cutoff(mean_user,std_user)
    grade_printer(marks_of_user)
elif (inp == 2):
    df = pd.read_csv(link_csv)
    for col in df.columns:
        if col.lower() == 'marks':
            mean_df = df[col].mean()
    for col in df.columns:
        if col.lower() == 'std':
            std_df = df[col].std()
    marks_df_input = float(input("Enter your marks to calculate your grade:"))
    grading_cutoff(mean_df,std_df)
    grade_printer(marks_df_input)
elif(inp == 3):
    mean_array = np.mean(marks_array)
    std_array = np.std(marks_array)
    grading_cutoff(mean_array, std_array)
    marks_input_array = float(input("Enter your marks to calculate your grade:"))
    grade_printer(marks_input_array)

    
    
```
